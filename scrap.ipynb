## Sword image collection

This ipynb file aim to find as much sword image from museum api and website scrapping as possible

```python
import requests
import pandas as pd

#The metropolitan museum

df = pd.read_csv("file_for_scrap/MetObjects.csv")

df.columns
```

```python
sword_df = df[df["Object Name"].str.contains("sword|katana|rapier|scimitar|blade", case=False, na=False)]
import requests
from bs4 import BeautifulSoup

def get_image_url(museum_page_url):
    response = requests.get(museum_page_url)
    soup = BeautifulSoup(response.text, "html.parser")
    image_tag = soup.find("meta", property="og:image")
    if image_tag:
        print("get!")
        return image_tag["content"]
    return None

sword_df["Image URL"] = sword_df["Link Resource"].apply(get_image_url)
print(sword_df[["Title", "Object Name", "Culture", "Image URL"]].head())
```

```python
import os

# Create directory for images
os.makedirs("met_sword_images", exist_ok=True)

# Download images
for index, row in sword_df.iterrows():
    img_url = row["Image URL"]
    
    if pd.notna(img_url):
        img_data = requests.get(img_url).content
        filename = f"sword/{row['Object ID']}.jpg"
        
        with open(filename, "wb") as file:
            file.write(img_data)

print("Downloaded all sword images from Met Museum!")
```
